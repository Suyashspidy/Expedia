```python
#Load the required libraries
import pandas as pd
import numpy as np
import seaborn as sns

#Load the data
df = pd.read_excel('/data10k.xlsx')


#View the data
df.head()
```

```python
#Basic information

df.info()

#Describe the data

df.describe()
```

```python
#Find null values

df.isnull().sum()
```

```python
median_value = df["orig_destination_distance"].median()
df["orig_destination_distance"].fillna(median_value, inplace=True)
median_value = df["user_location_longitude"].median()
df["user_location_longitude"].fillna(median_value, inplace=True)
median_value = df["user_location_latitude"].median()
df["user_location_latitude"].fillna(median_value, inplace=True)
median_value = df["srch_ci"].median()
df["srch_ci"].fillna(median_value, inplace=True)
median_value = df["srch_co"].median()
df["srch_co"].fillna(median_value, inplace=True)
mode_value = df["user_location_region"].mode()[0]
df["user_location_region"].fillna(mode_value, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.to_csv("data_with_imputed_values.csv", index=False)
```
