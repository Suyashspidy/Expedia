```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
df = pd.read_csv("Final_Final_Matrix.csv",encoding="ISO-8859-1")
```

```python
df.iloc[:,7:14]
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Ordinary Least Squares (OLS) model
Activity = ols('Activity ~ Continent', data=df).fit()
anova_table = sm.stats.anova_lm(Activity, typ=2)
print(anova_table)

pair_t = Activity.t_test_pairwise('Continent')
pair_t.result_frame
```
